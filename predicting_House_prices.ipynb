```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400])
y = np.array([300000, 320000, 340000, 360000, 380000, 400000, 420000, 440000, 460000, 480000])


a = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis]
pinv = np.linalg.pinv(a)
alpha = pinv.dot(y)


plt.plot(x, y, 'bo', label="x=house size, y=price")
plt.plot(x, alpha[0] * x + alpha[1], 'r-', label="least squares regression")
plt.legend()
plt.xlabel('House Size (sqft)')
plt.ylabel('House Price ($)')
plt.title('House Size vs Price')
plt.show()

# Print the parameters
print(f'Intercept: {alpha[1][0]}, Slope: {alpha[0][0]}')
```
