```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
y = [0, 0.8, 0.9, 0.2, -0.8, -1, -0.8, -0.4, -0.2]

f = CubicSpline(x,y)

new_x = np.linspace(0,2,100)
new_y = f(new_x)

plt.plot(x,y,"r-",label = "titik lama")
plt.plot(new_x,new_y,"b-",label = "titik baru")
plt.title("Cubic Spline Interpolation")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```
